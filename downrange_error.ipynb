# **Down Range error and overshoot of rockets**

```python
import numpy as np
import math
import ipywidgets as widgets
from IPython.display import display

print("Calculating the overshoot of a rocket due to its initial conditions")

# Define the sliders
vbo_slider = widgets.FloatSlider(min=0.5, max=1.5, step=0.001, value=0.905, description='Vbo:')
rbo_slider = widgets.FloatSlider(min=0.5, max=1.5, step=0.001, value=1.1, description='Rbo:')
phibo_slider = widgets.FloatSlider(min=0, max=90, step=1, value=30, description='Phibo:')
roc_vbo_slider = widgets.FloatSlider(min=-0.0005, max=0.0005, step=0.00001, value=-0.00005, description='Burnout error for V:')
roc_rbo_slider = widgets.FloatSlider(min=-0.005, max=0.005, step=0.0001, value=0.0005, description='Burnout error for R:')
roc_phibo_slider = widgets.FloatSlider(min=-0.0005, max=0.0005, step=0.00001, value=-0.0001, description='Burnout error for Phi:')
psi_slider = widgets.FloatSlider(min=0, max=180, step=1, value=100, description='Initial launch angle:')
distance_slider = widgets.FloatSlider(min=1000, max=10000, step=1, value=3444, description='Distance to travel:')

# Define the output widget
output = widgets.Output()

# Define the function to calculate the overshoot
def calculate_overshoot(change):
    
    with output:
        # Get the current slider values
        vbo = vbo_slider.value
        rbo = rbo_slider.value
        phibo = phibo_slider.value
        roc_vbo = roc_vbo_slider.value
        roc_rbo = roc_rbo_slider.value
        roc_phibo = roc_phibo_slider.value
        psi = psi_slider.value
        distance = distance_slider.value
        
        # Calculate the overshoot
        drbo = (4*(np.sin(math.radians((psi/2)))**2))/((vbo**2)*(rbo**2)*(np.sin(math.radians(phibo*2))))
        dvbo = (2*(rbo)*drbo) / (vbo)
        dphibo = ((2 * (np.sin(math.radians(psi+(2*phibo))))) / (np.sin(math.radians(phibo*2)))) - 2
        down_range = (drbo*roc_rbo) + (dvbo*roc_vbo) + (dphibo*roc_phibo)
        overshoot = down_range * distance

        # Print the results
        print(f"drbo: {drbo}")
        print(f"dvbo: {dvbo}")
        print(f"dphibo: {dphibo}")
        print(f"Down range error: {down_range}")
        print(f"Overshoot: {round(overshoot,2)} n.mi. overshoot (rounded to 2 decimal points)")

# Call the function initially to display the default values
calculate_overshoot(None)

# Observe changes in the sliders and update the output
vbo_slider.observe(calculate_overshoot, 'value')
rbo_slider.observe(calculate_overshoot, 'value')
phibo_slider.observe(calculate_overshoot, 'value')
roc_vbo_slider.observe(calculate_overshoot, 'value')
roc_rbo_slider.observe(calculate_overshoot, 'value')
roc_phibo_slider.observe(calculate_overshoot, 'value')
psi_slider.observe(calculate_overshoot, 'value')
distance_slider.observe(calculate_overshoot, 'value')


display(vbo_slider)
display(rbo_slider)
display(phibo_slider)
display(roc_vbo_slider)
display(roc_rbo_slider)
display(roc_phibo_slider)
display(psi_slider)
display(distance_slider)
display(output)
```
